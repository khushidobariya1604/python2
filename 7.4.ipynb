# Air Polution

```python
import requests,json

def weather_data():
    api_key = 'f14acf72dcf9e30d8319bb30f98bde38'
    request_url = f'http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}'
    result=requests.get(request_url)
    wd=result.json()
    print(json.dumps(wd,indent=5)) 
    aqi=wd['list'][0]['main']['aqi']
    quality_index={1:'Good',2:'Fair',3:'Moderate',4:'Poor',5:'Very Poor'}
    print("Air Quality Index = ",quality_index[aqi])

weather_data()
```

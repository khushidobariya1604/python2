```python
d=pd.read_csv("py.csv")
x=d.iloc[:,0:1]
y=d.iloc[:,-1]
```

```python
# Split the data in training ,Testing

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
error=[]

for i in range(0,42):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    error.append(metrics.mean_absolute_error(y_test,y_pred))
print(error)
print(min(error))
```

```python
error_dict={}

for i in range(0,42):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    error_dict[i]=metrics.mean_absolute_error(y_test,y_pred)
print(error_dict)
```

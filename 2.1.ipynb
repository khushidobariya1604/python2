# Data Visualization

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=np.random.randint(1,100,size=(10,10))
print(data)
```

```python
sns.heatmap(data)
plt.show()
```

```python
sns.heatmap(data,cmap='tab20')
plt.show()
```

```python
sns.heatmap(data,vmin=30,vmax=70)
plt.show()
```

```python
sns.heatmap(data,annot=True,cbar=False)
plt.show()
```

```python
sns.heatmap(data,annot=True)
plt.show()
```

```python
sns.heatmap(data,annot=True,cbar=False,linewidths=2)
plt.show()
```

```python
sns.heatmap(data,annot=True,cbar=False,linewidths=2,linecolor='yellow')
plt.show()
```

```python
sns.heatmap(data,annot=True,cbar=False,linewidths=2,linecolor='black',xticklabels=False,yticklabels=False)
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(np.random.random((5,5)),
                columns=['a','b','c','d','e'])
df
```

```python
sns.heatmap(df)
plt.show()
```

```python
df=pd.DataFrame(np.random.random((3,10)),
                columns=['a','b','c','d','e','f','g','h','i','j'])
df
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(df,cmap='tab20',linewidths=2,linecolor='black')
plt.show()
```

```python
df=pd.read_csv('auto-mpg.csv')
sns.boxplot(y=df['mpg'],data=df)
plt.show()
```

```python
df=pd.read_csv('auto-mpg.csv')
sns.boxplot(y=df['acceleration'],data=df)
plt.show()
```

```python
df=pd.read_csv('auto-mpg.csv')
sns.boxplot(x=df['mpg'],y=df['acceleration'],data=df,hue='cylinders')
plt.show()
```

```python
df=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df)
plt.show()
```

**- Set_index() :**
- used to set one or more columns of a dataframe as the index

```python
data={'name':['alice','bob','charlie'],
      'age':[25,30,35],
      'city':['nyc','la','chicago']}
df=pd.DataFrame(data)
df.set_index('name',drop=False)
```

```python
df
```

```python
df.set_index('name',drop=False,inplace=True)
```

```python
df
```

```python
df1=df.set_index(['name','age'],drop=False)
```

```python
df1
```

**- reset_index() :**
- used to manage and reset the index of a dataframe

```python
data={'name':['alice','bob','charlie'],
      'age':[25,30,35],
      'city':['nyc','la','chicago']}
f1=pd.DataFrame(data,index=['a','b','c'])
f1
```

```python
f_reset=f1.reset_index()
f_reset
```

```python
f_reset=f1.reset_index(drop=True)
f_reset
```

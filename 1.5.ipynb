# Drop and Remove duplicates

```python
import pandas as pd
```

```python
data={'a':['TeamA','TeamB','TeamB','Teamc','TeamA'],
      'b':[50,40,40,30,50],
       'c':[True,False,False,False,True]}
df=pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
df.duplicated(keep='first') # CHeck from first
```

```python
df.duplicated(keep='last') # CHeck from last
```

```python
df.duplicated(keep=False) # gives True in all duplicates
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep='first')
```

```python
df.drop_duplicates(keep='last')
```

```python
df.drop_duplicates(keep=False)
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.reset_index(drop=True)
```

# apply()

```python
s=pd.Series([35,67,90,45])
def pass_fail(x):
    return "pass" if x>=50 else "fail"

result=s.apply(pass_fail)
print(result)
```

```python
df=pd.DataFrame({
    'Physics':[60,80,80],
    'Chemistry':[65,75,85],
    'Maths':[70,80,90]
})
df
```

```python
result=df.apply(sum)
print(result)
```

```python
result=df.apply(sum,axis=1)
print(result)
```

```python
df=pd.DataFrame({
    'Physics':[60,70,80],
    'Chemistry':[65,75,85],
    'Maths':[70,80,90]
})

result=df.apply(lambda x:x.mean())
print(result)
```

```python
df
```

```python
df['total']=df.apply(lambda x:x.sum(),axis=1)
df
```

```python
def grade(total):
    if total>=250:
        return 'A'
    elif total>=200:
        return 'B'
    else:
        return 'C'
```

```python
df['grade']=df['total'].apply(grade)
df
```

## Raw parameter :
- less time consuming
- pass the data as an array

```python
df=pd.DataFrame({
    'width':[1,2],'height':[3,4]
})

def get_stats(row):
    return [row['width']*row['height'],2*(row['width']+row['height'])]
df
```

```python
df.apply(get_stats,axis=1)
```

```python
df.apply(get_stats,axis=1,result_type='expand')
```

```python
df.apply(get_stats,axis=1,result_type='reduce')
```

```python
df.apply(get_stats,axis=1,result_type='broadcast')
```

# Args and Kwargs parameters

```python
def add_values(x,y,z):
    return x+y+z

df=pd.DataFrame({'A':[1,2],'B':[10,20]})

d=df.apply(add_values,args=(1,2))

print(d)
```

```python
def multiply_by_m(x,m):
    return x*m

df=pd.DataFrame({'A':[1,2],'B':[10,20]})

d=df.apply(multiply_by_m,m=2)

print(d)
```

### replace() function

```python
df=pd.DataFrame({
    'Marks':[80,85,0,90],
    'Attendance':[75,80,0,85],
    'Grade':['A','B','F','D']
})
print(df)
result=df.replace([80,75],[90,80])
print(result)
```

```python
df=pd.DataFrame({
    'Marks':[80,85,0,90],
    'Attendance':[75,80,0,85],
    'Grade':['A','B','F','D']
})
result=df.replace({80:90},{85:80})
print(result)
```

```python
df=pd.DataFrame({
    'Marks':[80,85,0,90],
    'Attendance':[75,80,0,85],
    'Grade':['A','B','F','D']
})
print(df)
result=df.replace(0,95)
print(result)
```

```python
df['Marks']=df['Marks'].replace(80,85)
df
```

## concat() function

```python
d1=pd.DataFrame({'A':[1,2],'B':[3,4]})
d2=pd.DataFrame({'A':[5,6],'C':[7,8]})

r=pd.concat([d1,d2],axis=0)
r
```

```python
d1=pd.DataFrame({'A':[1,2],'B':[3,4]})
d2=pd.DataFrame({'A':[5,6],'C':[7,8]})

r=pd.concat([d1,d2],axis=1)
r
```

```python
r=pd.concat([d1,d2],join='outer') # keeps all col and rows
r
```

```python
r=pd.concat([d1,d2],join='inner') # only keeps common/overlap col and rows
r
```

```python
r=pd.concat([d1,d2],join='inner',ignore_index=True) 
r
```

```python
d1=pd.DataFrame({'A':[1,2],'B':[3,4]})
d2=pd.DataFrame({'C':[5,6],'D':[7,8]})

r=pd.concat([d1,d2],sort=True) 
r
```

```python
r=pd.concat([d1,d2],sort=False) 
r
```

```python
r=pd.concat([d1,d2],sort=False,axis=1) 
r
```

# merge()  function

```python
d1=pd.DataFrame({
    'Employee':['A' , 'B' , 'C'],
    'Department':['IT','HR','Finance']
})

d2=pd.DataFrame({
    'Employee':['A' , 'B' , 'D'],
    'salary':[50000,40000,30000]
})

r=pd.merge(d1,d2,on='Employee',how='inner')
print(r)
```

```python
r=pd.merge(d1,d2,on='Employee',how='outer')
print(r)
```

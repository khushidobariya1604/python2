```python
import pandas as pd
```

```python
a1=pd.DataFrame({
    'id':[1,2,3],
    'name':['amit','riya','john'],
    'salary':[50000,60000,55000]
})
a2=pd.DataFrame({
    'id':[1,2,4],
    'name':['amit','riya','dravid'],
    'salary':[50000,60000,58000]
})

r=pd.merge(a1,a2,on='id',suffixes=("_a1","_a2"))
print(r)
```

- how="inner" --> gives common elements
- how="outer" --> gives all elements

```python
d1=pd.DataFrame({
    "student_id":[1,2,3],
    "name":['Amit','Riya','John']
})
d2=pd.DataFrame({
    "id":[1,2,4],
    "marks":[85,90,75]
})
r=pd.merge(d1,d2,left_on='student_id',right_on='id')
print(r)
```

```python
d1=pd.DataFrame({
    "physics":[70,80,90]},
    index=[101,102,103]
)
d2=pd.DataFrame({
    "name":['alice','bob','charlie']},
     index=[101,102,103])
    
r=pd.merge(d1,d2,left_index=True,right_index=True)
print(r)
```

```python
d1=pd.DataFrame({
    "student_id":[1,2,3],
    "name":['Amit','Riya','John']
})
d2=pd.DataFrame({
    "score":[80,90,70]},
     index=[1,2,3])

r=pd.merge(d1,d2,left_on='student_id',right_index=True)
print(r)
```

# groupby() function

```python
data={'product':['electronics','furniture','electronics','furniture'],
     'sales':[200,150,300,100]}
df=pd.DataFrame(data)
```

```python
r=df.groupby('product')['sales'].sum()
print(r)
```

```python
r=df.groupby('product')['sales'].mean()
print(r)
```

```python
r=df.groupby('product')['sales'].median()
print(r)
```

```python
r=df.groupby('product')['sales'].describe()
print(r)
```

```python
r=df.groupby('product')['sales'].sample()
print(r)
```

```python
r=df.groupby('product')['sales'].size()
print(r)
```

```python
r=df.groupby('product')['sales'].groups
print(r)
```

```python
r=df.groupby('product')['sales'].nunique()
print(r)
```

```python
r=df.groupby('product')['sales'].nth(0)
print(r)
```

```python
r=df.groupby('product')['sales'].nth(1)
print(r)
```

```python
r=df.groupby('product',axis=0,as_index=True,sort=True)['sales'].count()
print(r)
```

```python
r=df.groupby('product',axis=0,as_index=False,sort=True)['sales'].count()
print(r)
```

```python
df=pd.DataFrame({
    'Region':['East','East','West','West','East','West'],
    'sales':[100,150,200,120,110,130],
    'Manager':['A','B','C','D','E','F']
})
df
```

```python
def sales_range(group):
    return group['sales'].max()-group['sales'].min()
sales=df.groupby('Region').apply(sales_range)
print(sales)
```

```python
df=pd.DataFrame({
    'dept':['sales','sales','HR','HR','IT','IT'],
    'salary':[50000,60000,40000,45000,70000,75000],
    'bonus':[5000,6000,2000,2500,8000,5800]
})
r=df.groupby('dept').agg('mean')
print(r)
```

```python
df=pd.DataFrame({
    'dept':['sales','sales','HR','HR','IT','IT'],
    'salary':[50000,60000,40000,45000,70000,75000],
    'bonus':[5000,6000,2000,2500,8000,5800]
})
r=df.groupby('dept').agg(['mean','min','max'])
print(r)
```

```python
df=pd.DataFrame({
    'dept':['sales','sales','HR','HR','IT','IT'],
    'salary':[50000,60000,40000,45000,70000,75000],
    'bonus':[5000,6000,2000,2500,8000,5800]
})
r=df.groupby('dept').agg({'salary':['mean','max'],
                          'bonus':['sum']})
print(r)
```

```python
df=pd.DataFrame({
    'dept':['sales','sales','HR','HR','IT','IT'],
    'salary':[50000,60000,40000,45000,70000,75000],
    'bonus':[5000,6000,2000,2500,8000,8500]
})
r=df.groupby('dept').agg(Meansalary=('salary','mean'),
                        maxsalary=('salary','max'),
                        totalbonus=('bonus','sum'))
print(r)
```

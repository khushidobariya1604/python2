```python
import pandas as pd
```

```python
df=pd.read_csv('book1.csv')
df
```

```python
df.info()
```

- cgpa = x
- package = y (predicted)

```python
x=df.iloc[:,0:1]  # or x=df[["cgpa"]]
y=df.iloc[:,-1]   # or y=df["package"]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)  # random_state is used for getting accurate value of train,test

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
a=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
a
```

```python
print("Coefficient(m) = ",lr.coef_)
print("Intercept(c) = ",lr.intercept_)
```

```python
from sklearn import metrics

print("MAE : ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE : ",metrics.mean_squared_error(y_test,y_pred))
print("R2 score : ",metrics.r2_score(y_test,y_pred))
```

```python
from sklearn import metrics
random_state=42
for i in range(0,random_state):
    print(i)
    print("MAE : ",metrics.mean_absolute_error(y_test,y_pred))
    print("MSE : ",metrics.mean_squared_error(y_test,y_pred))
    print("R2 score : ",metrics.r2_score(y_test,y_pred))
    print()
```

# Multiple Linear Regression


```python
df=pd.read_csv('cardata.csv')
df
```

```python
df.info()
```

```python
# Drop Carname column
df.drop('Car_Name',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Age']=2026-df['Year']
```

```python
df.columns
```

```python
df.head(2)
```

```python
df.drop('Year',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df = pd.get_dummies(data=df,drop_first=True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
x=df.drop('Selling_Price',axis=1) # or x=df.iloc[:,1:]
y=df['Selling_Price']  # or y=df.iloc[:,0]
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)  # random_state is used for getting accurate value of train,test

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(y_pred)
```

```python
a=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
a
```

```python
print("Coefficient(m) = ",lr.coef_)
print("Intercept(c) = ",lr.intercept_)
```

```python
from sklearn import metrics

print("MAE : ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE : ",metrics.mean_squared_error(y_test,y_pred))
print("R2 score : ",metrics.r2_score(y_test,y_pred)
```

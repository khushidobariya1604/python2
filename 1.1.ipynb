**> Pandas Series :**

- series is like a column in table. it is 1-D array holding data of any type.

```python
import pandas as pd
a=[1,7,2]
var=pd.Series(a)
print(var)
print(type(var))
```

```python
print(var[1])
```

change the name of index

```python
import pandas as pd
a=[1,7,2]
var=pd.Series(a,index=['x','y','z'])
print(var)
print(type(var))
```

```python
print(var['y'])
```

give input as tuple

```python
import pandas as pd
a=(1.4,2,3,4)
var=pd.Series(a)
print(var)
print(type(var))
```

```python
import pandas as pd
s_name=['ram','shyam','radha','geeta','seeta']
marks=[50,60,70,80,90]
var=pd.Series(marks,index=s_name,name='Student Result')
print(var)
```

```python
import pandas as pd
calories={'day1':400,'day2':380,'day3':430,'day4':390}
var=pd.Series(calories)
print(var)
```

```python
import pandas as pd
calories={'day1':400,'day2':380,'day3':430,'day4':390}
var=pd.Series(calories,index=['day1','day2'])
print(var)
```

```python
import pandas as pd
calories={'day1':400,'day2':380,'day3':430,'day4':390}
var=pd.Series(calories,name='Calories Burn per day')
print(var)
```

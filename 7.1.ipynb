```python
url='https://subslikescript.com/movies'
result=requests.get(url)
content=result.text
```

```python
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
len(soup.find_all('a'))
```

```python
len(soup.find_all('a',href=True))
```

```python
box=soup.find('article',class_='main-article')
print(box)
```

```python
box=soup.find('article',class_='main-article')
print(box)
```

```python
links=[]
for l in box.find_all('a',href=True):
    print(type(l))
    links.append(l['href'])
print(links)
```

```python
urls=[]
for i in links:
    urls.append('https://subslikescript.com/movie'+i)
```

```python
print(urls)
```

```python
for i in urls:
    result=requests.get(i)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    box=soup.find('article',class_='main-article')
    title=box.find('h1').get_text()
    script=box.find('div',class_='full-script').get_text()
    try:
        with open(f"{title}.txt",'w',encoding='utf-8') as file:
            file.write(script)
    except Exception as e:
        print(e)
```

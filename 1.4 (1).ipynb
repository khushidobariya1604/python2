# outliers



- low_value = Q1 - 1.5 * IQR
- high_value = Q3 + 1.5 * IQR

```python
def find_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    
    IQR=q3-q1
    print(IQR)
    
    low_val=q1-1.5*IQR
    print(low_val)
    
    high_val=q3+1.5*IQR
    print(high_val)
    
    ds=ds.loc[(ds[col]<low_val) | (ds[col]>high_val)]
    
    return ds

```

```python
df=pd.read_csv('auto-mpg.csv')
find_outliers(df,'mpg')
```

```python
find_outliers(df,'acceleration')
```

```python
def remove(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    
    IQR=q3-q1    
    low=q1-1.5*IQR
    print(low)
    high=q3+1.5*IQR
    print(high)
    
    ds=ds.loc[(ds[col]>=low) & (ds[col]<=high)]
    
    return ds
```

```python
df=pd.read_csv('auto-mpg.csv')
remove(df,'mpg')
```

```python
remove(df,'acceleration')
```
